## **3. DASAR TEORI**

**Variabel** adalah lokasi memori untuk menyimpan data sementara. Python menggunakan dynamic typing, sehingga tipe data ditentukan otomatis dari nilai yang diberikan.

**Tipe Data** dasar Python: int (bilangan bulat), float (desimal), str (teks), bool (logika True/False).

**Percabangan** (if–elif–else) digunakan untuk pengambilan keputusan berdasarkan kondisi tertentu. Operator yang digunakan: >, <, >=, <=, ==, !=, and, or, not.

**Perulangan** digunakan untuk mengeksekusi blok kode berulang kali. for digunakan saat jumlah iterasi diketahui, while digunakan saat iterasi bergantung pada kondisi.

**Fungsi** adalah kumpulan kode yang bisa digunakan berulang kali. Fungsinya membantu membuat program lebih terstruktur. Fungsi dibuat menggunakan kata kunci def.

**List** adalah struktur data yang menyimpan banyak nilai dalam satu variabel, bersifat ordered dan mutable. Operasi umum: append(), remove(), len(), max(), min().

### **TUGAS 1 (Variabel dan Tipe Data)**


```python
nama = input("Masukkan nama: ")
umur = int(input("Masukkan umur: "))
tinggi = float(input("Masukkan tinggi badan (cm): "))

print("\n=== DATA DIRI ===")
print(f"Nama   : {nama}")
print(f"Umur   : {umur} tahun")
print(f"Tinggi : {tinggi} cm")
```

**Analisis Program**:

Program menerima tiga input dengan tipe data berbeda, yaitu string (nama), integer (umur), dan float (TB). Data tersebut kemudian ditampilkan menggunakan f-string. Fungsi int() dan float() dipakai untuk mengubah input yang awalnya berupa string menjadi tipe data yang sesuai.

### **TUGAS 2 (Percabangan)**





```python
# kelulusan

nilai = int(input("Masukkan nilai: "))

if nilai >= 60:
    print("Status: Lulus")
else:
    print("Status: Tidak Lulus")
```

**Analisis Program**:

Program menggunakan struktur if-else untuk membandingkan nilai dengan batas kelulusan 60. Jika input nilai ≥ 60, bagian if dijalankan. Jika imput nilai kurang dari 60 bagian else yang dijalankan.

### **TUGAS 3 (PERULANGAN FOR)**

```python
# perulangan angka 1-20

print("Angka 1 sampai 20:")
for i in range(1, 21):
    print(i)
```

**Analisis Program**:

Range(1, 21) menghasilkan angka dari 1 sampai 20.Tiap angka yang dicetak (1-20) ditampilkan di baris baru.


### **TUGAS 4 (PERULANGAN + PERCABANGAN)**

```python
# perulangan + percabangan (bil genap 1-50)

print("Bilangan genap dari 1 sampai 50:")
for i in range(1, 51):
    if i % 2 == 0:
        print(i)
print()
```

**Analisis Program**:

Operator modulo % digunakan untuk mengecek apakah bilangan genap, yaitu jika habis dibagi 2. Perulangan for dan kondisi if digunakan untuk memilih angka tertentu dalam suatu rentang.

### **TUGAS 5 (LIST DASAR)**

```python
angka = []

for i in range(5):
    nilai = int(input("Masukkan angka: "))
    angka.append(nilai)

print("\nIsi list:")
for item in angka:
    print(item)
```

**Analisis Program**:

Program membuat list lalu mengisi 5 angka dari input pengguna. Setiap angka disimpan dengan append(), kemudian semua isi list ditampilkan menggunakan perulangan for.

### **TUGAS 6 (LIST + PERULANGAN)**

```python
# list + perulangan input 5 nilai mahasiswa

nilai = []

for i in range(1, 6):
    n = int(input(f"Masukkan nilai mahasiswa ke-{i}: "))
    nilai.append(n)

total = sum(nilai)
rata_rata = total / len(nilai)

print(f"\nData nilai  : {nilai}")
print(f"Total nilai : {total}")
print(f"Rata-rata   : {rata_rata:.2f}")
```

**Analisis Program**:

Program menyimpan 5 nilai ke dalam list menggunakan perulangan dan append(). Data kemudian dijumlahkan dengan sum() dan dihitung rata-ratanya dengan membagi total dengan jumlah data. Hasil ditampilkan dengan format dua angka di belakang koma.

### **TUGAS 7 (FUNGSI)**



```python
# fungsi luas persegi panjang

def hitung_luas(panjang, lebar):
    return panjang * lebar

panjang = float(input("Masukkan panjang: "))
lebar   = float(input("Masukkan lebar  : "))

luas = hitung_luas(panjang, lebar)
print(f"Luas persegi panjang: {luas}")
```

**Analisis Program**:

Program membuat fungsi hitung_luas() untuk menghitung luas persegi panjang dari panjang dan lebar. Hasil perhitungan disimpan dalam variabel luas.

### **TUGAS 8 (PERCABANGAN BERTINGKAT)**

```python
# percabangan bertingkat (kategori nilai)

nilai = int(input("Masukkan nilai: "))

if nilai >= 85:
    grade = "A"
elif nilai >= 70:
    grade = "B"
elif nilai >= 60:
    grade = "C"
else:
    grade = "D"

print(f"Grade: {grade}")
```

**Analisis Program**:

Program menerima input nilai lalu menentukan grade dengan percabangan if-elif-else. Setiap kondisi mewakili rentang nilai tertentu. Hasilnya disimpan dalam variabel grade.

### **TUGAS 9 (PERULANGAN WHILE)**

```python
# perulangan while angka 1–10

i = 1
while i <= 10:
    print(i)
    i += 1
print()
```

**Analisis Program**:

Program menggunakan perulangan while untuk menampilkan angka dari 1 sampai 10. Variabel i dimulai dari 1 hingga 10.

### **TUGAS 10 (LIST + PENCARIAN)**

```python
# list + pencarian nilai terbesar & terkecil

angka = []
n = int(input("Jumlah banyak angka yang mau diinput: "))

for i in range(1, n + 1):
    a = int(input(f"Masukkan angka ke-{i}: "))
    angka.append(a)

print(f"\nData   : {angka}")
print(f"Terbesar : {max(angka)}")
print(f"Terkecil : {min(angka)}")
```

**Analisis Program**:

Program membuat list dan mengisi sejumlah angka sesuai input pengguna dengan perulangan dan append(). Data kemudian ditampilkan lalu nilai terbesar dicari dengan max() dan nilai terkecil dengan min().

### **TUGAS 11 (GABUNGAN SEMUA KONSEP)**

```python
def hitung_rata_rata(data):
    return sum(data) / len(data)

nilai = []

for i in range(1, 6):
    n = int(input(f"Masukkan nilai mahasiswa ke-{i}: "))
    nilai.append(n)

rata = hitung_rata_rata(nilai)

if rata >= 60:
    status = "Lulus"
else:
    status = "Tidak Lulus"

print(f"\nData nilai  : {nilai}")
print(f"Rata-rata   : {rata:.2f}")
print(f"Status      : {status}")
```

**Analisis Program**:

Program membuat fungsi hitung_rata_rata() untuk menghitung rata-rata data nilai. Nilai dimasukkan ke dalam list dengan perulangan. Hasil rata-rata digunakan untuk menentukan status lulus atau tidak dengan percabangan if-else

### **TUGAS 12 (SIMULASI KASIR)**

```python
harga_barang = []
n = int(input("Jumlah banyak barang yang mau dibeli: "))

for i in range(1, n + 1):
    harga = float(input(f"Masukkan harga barang ke-{i}: Rp "))
    harga_barang.append(harga)

total = sum(harga_barang)
diskon = 0

if total >= 100000:
    diskon = total * 0.10
    print("\nAnda mendapat diskon 10%")

total_bayar = total - diskon

print(f"\nDaftar harga : {harga_barang}")
print(f"Total belanja: Rp {total:,.0f}")
print(f"Diskon       : Rp {diskon:,.0f}")
print(f"Total bayar  : Rp {total_bayar:,.0f}")
```

**Analisis Program**:

Program meminta memasukkan harga beberapa barang, lalu menyimpannya ke dalam list. Program menjumlahkan semua harga untuk mendapatkan total belanja. Jika total mencapai 100000 atau lebih, mendapat diskon 10%. Terakhir, program menghitung total yang harus dibayar.

### **TUGAS 13 (TABEL PERKALIAN DINAMIS)**

```python
angka = int(input("Masukkan angka: "))

print(f"\nTabel Perkalian {angka}:")
for i in range(1, 11):
    hasil = angka * i
    print(f"  {angka} x {i:2} = {hasil:3}")
```

### **Analisis Program**:

Program menerima satu angka, lalu menampilkan tabel perkalian dari 1 sampai 10 dengan perulangan for.

### **TUGAS 14 (FAKTORIAL DENGAN FUNGSI)**

```python
def faktorial(n):
    hasil = 1
    for i in range(1, n + 1):
        hasil *= i
    return hasil

angka = int(input("Masukkan angka: "))
print(f"Faktorial dari {angka} adalah {faktorial(angka)}")
```

**Analisis Program**:

Program membuat fungsi faktorial() untuk menghitung hasil perkalian berurutan dari 1 sampai n dengan perulangan for.

### **TUGAS 15 (LOGIN SEDERHANA DENGAN PERULANGAN**

```python
username_benar = "admin"
password_benar = "123"

kesempatan = 3

for i in range(kesempatan):
    username = input("Masukkan username: ")
    password = input("Masukkan password: ")

    if username == username_benar and password == password_benar:
        print("Login berhasil")
        break
    else:
        print("Username atau password salah")

else:
    print("Akses ditolak")
```

**Analisis Program**:

Program menggunakan perulangan for untuk memberi 3 kali kesempatan login. Menggunakan perulangan if-else untuk mengecek username dan password. Jika benar, program berhenti. Jika salah terus sampai habis, program menampilkan “Akses ditolak”.
