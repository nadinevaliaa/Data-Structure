## **3. DASAR TEORI**

Array dan list merupakan struktur data linear yang menyimpan sekumpulan elemen data secara berurutan dalam satu variabel. Pada Python, struktur data linear yang paling fleksibel dan sering digunakan adalah List. Karakteristik utama list di Python meliputi kemampuannya menyimpan banyak data dengan tipe berbeda, datanya tersusun berurutan menggunakan indeks (dimulai dari 0), serta bersifat mutable (nilainya dapat diubah).

Operasi-operasi dasar pada list meliputi:
1. Insert (Penyisipan): Menambahkan data baru, contohnya menggunakan fungsi .append() untuk menambah di akhir list.  
2. Delete (Penghapusan): Menghapus elemen tertentu menggunakan fungsi .remove().
3. Update (Pembaruan): Mengubah nilai elemen pada indeks spesifik.  
4. Traversal (Penelusuran): Membaca dan mengakses seluruh elemen list satu per satu secara berurutan menggunakan perulangan (looping).  
5. Searching (Pencarian): Menemukan posisi atau keberadaan data. Metode paling dasar adalah Linear Search dengan memeriksa elemen dari awal hingga akhir dengan kompleksitas $O(n)$.
5. Sorting (Pengurutan): Menyusun data secara terurut baik membesar (ascending) maupun mengecil (descending).

```python
# Tugas 5: Mengurutkan Data Ascending dan Descending

nilai = [34, 12, 89, 5]
nilai.sort()
print("Ascending:", nilai)
nilai.sort(reverse=True)
print("Descending:", nilai)
```

```python
# Tugas 6: Menentukan Nilai Terbesar dan Terkecil
skor = [78, 92, 45, 85]
print("Terbesar:", max(skor))
print("Terkecil:", min(skor))
```

```python
# Tugas 7: Pengolahan Nilai Mahasiswa

nilai_mhs = []
for i in range(5):
    nilai_mhs.append(float(input(f"Nilai {i+1}: ")))
print("Rata-rata:", sum(nilai_mhs)/len(nilai_mhs))
print("Tertinggi:", max(nilai_mhs))
```

### Analisis:

Penggunaan keyword in pada Python mempermudah pencarian linear tersembunyi pada list string tanpa perlu menulis struktur loop manual secara panjang.

```python
# Tugas 9: Sorting Manual Tanpa Menggunakan sort()

data = [64, 34, 25, 12]
n = len(data)
for i in range(n):
    for j in range(0, n-i-1):
        if data[j] > data[j+1]:
            data[j], data[j+1] = data[j+1], data[j]
print("Hasil urut manual:", data)
```

### Analisis:

Algoritma Bubble Sort manual ini membandingkan dua elemen yang bersebelahan secara berulang lalu menukarnya jika urutannya salah sampai seluruh list terurut.
