## **DISKUSI**

**a. Apa perbedaan array dan list?**  

* **Array:** Biasanya memiliki ukuran yang tetap sejak awal dideklarasikan dan umumnya hanya bisa menyimpan elemen dengan satu tipe data yang homogen.  

* **List:** Memiliki ukuran yang dinamis, membolehkan penambahan atau pengurangan elemen saat runtime, dan mampu menyimpan tipe data yang berbeda-beda dalam satu variabel.  

**b. Mengapa list disebut struktur data linear?**

Disebut linear karena elemen-elemen di dalamnya disusun dan disimpan secara berurutan dalam satu garis lurus memanjang, di mana setiap elemen terhubung secara sekuensial dengan elemen sebelum dan sesudahnya melalui sistem indeks berurutan.  

**c. Apa fungsi traversal?**

Fungsi utama traversal adalah untuk mengunjungi, membaca, atau memproses setiap elemen yang ada di dalam struktur data tepat satu kali secara berurutan, contohnya untuk menampilkan isi data ke layar atau melakukan operasi hitung kumulatif (seperti mencari rata-rata).  

**d. Mengapa searching penting dalam pengolahan data?**  

Searching sangat penting karena sistem pengolahan data membutuhkan kapabilitas untuk menemukan kembali atau memeriksa keberadaan suatu informasi spesifik di antara sekumpulan data yang besar secara cepat tanpa harus membaca keseluruhan data secara acak.

**e. Apa keuntungan sorting data?**  

Mengurutkan data membuat penyajian informasi menjadi lebih rapi dan mudah dibaca manusia. Selain itu, dari sudut pandang komputasi, data yang terurut merupakan prasyarat penting untuk mengoptimalkan algoritma pencarian lanjutan yang jauh lebih cepat, seperti Binary Search.
