```python
# Studi Kasus 1: Sistem Antrian Bank

from collections import deque

antrian = deque()

# Nasabah mengambil nomor antrian
antrian.append("San")
antrian.append("Mingi")
antrian.append("Yunho")

print("Antrian saat ini:", list(antrian))

# Memanggil nasabah
layani = antrian.popleft()
print("Nasabah yang dilayani:", layani)

print("Sisa antrian:", list(antrian))
```

### **Analisis:**

Program menerapkan konsep Queue dengan prinsip FIFO (First In First Out). Nasabah yang datang lebih dahulu akan dilayani terlebih dahulu. Operasi append() digunakan untuk menambah antrian dan popleft() digunakan untuk memanggil nasabah yang berada di depan antrian.

```python
# Studi Kasus 2: Sistem Pendaftaran Mahasiswa

from collections import deque

pendaftar = deque()

pendaftar.append("Upin")
pendaftar.append("Ijat")
pendaftar.append("Rambo")

print("Daftar pendaftar:", list(pendaftar))

diproses = pendaftar.popleft()
print("Mahasiswa diproses:", diproses)

print("Sisa pendaftar:", list(pendaftar))
```

### **Analisis:**

Mahasiswa yang mendaftar lebih awal akan diproses lebih dahulu. Queue sangat cocok digunakan karena urutan pelayanan harus sesuai urutan pendaftaran sehingga tidak terjadi ketidakadilan dalam proses registrasi.

```python
# Studi Kasus 3: Printer Queue

from collections import deque

printer = deque()

printer.append("Laprak.pdf")
printer.append("Tugas_ML.docx")
printer.append("SuratCinta.pdf")

print("Antrian cetak:", list(printer))

while printer:
    dokumen = printer.popleft()
    print("Mencetak:", dokumen)

print("Semua dokumen selesai dicetak")
```

### **Analisis:**

Printer mencetak dokumen berdasarkan urutan masuk. Dokumen pertama yang dikirim ke printer akan dicetak terlebih dahulu. Hal ini merupakan implementasi nyata dari Queue pada sistem operasi.

```python
# Studi Kasus 4: Call Center Koperasi

from collections import deque

telepon = deque()

telepon.append("Pelanggan A")
telepon.append("Pelanggan B")
telepon.append("Pelanggan C")

print("Antrian telepon:", list(telepon))

dilayani = telepon.popleft()
print("Sedang dilayani:", dilayani)

print("Sisa antrian:", list(telepon))
```

### **Analisis:**

Pelanggan yang menelepon lebih dahulu akan mendapatkan layanan lebih dahulu. Queue membantu menjaga urutan pelayanan sehingga proses call center menjadi lebih teratur dan efisien.

```python
# Studi Kasus 5: Buffer Streaming Video

from collections import deque

buffer_video = deque()

buffer_video.append("Frame 1")
buffer_video.append("Frame 2")
buffer_video.append("Frame 3")
buffer_video.append("Frame 4")

print("Isi buffer:", list(buffer_video))

while buffer_video:
    frame = buffer_video.popleft()
    print("Menampilkan:", frame)
```

### **Analisis:**

Data video diproses secara berurutan. Frame yang masuk lebih dahulu akan ditampilkan lebih dahulu sehingga video dapat diputar dengan urutan yang benar. Queue berperan sebagai buffer penyimpanan sementara data streaming.

```python
# Studi Kasus 6: Sistem Tiket Kereta Api Online

from collections import deque

pesanan = deque()

pesanan.append("TRX001 - Adit")
pesanan.append("TRX002 - Sopo")
pesanan.append("TRX003 - Jarwo")

print("Daftar pesanan:", list(pesanan))

while pesanan:
    proses = pesanan.popleft()
    print("Memproses:", proses)

print("Semua pesanan selesai diproses")
```

### **Analisis:**

Pesanan tiket diproses berdasarkan urutan transaksi yang masuk. Dengan Queue, sistem dapat menjamin bahwa pelanggan yang melakukan pemesanan lebih dahulu akan diproses terlebih dahulu, sehingga pelayanan menjadi adil dan konsisten.
