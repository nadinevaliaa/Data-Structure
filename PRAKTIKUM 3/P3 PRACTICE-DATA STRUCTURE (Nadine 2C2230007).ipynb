```python
# Praktikum 5: Perbandingan Efisiensi Algoritma

data = [10, 20, 30, 40]
target = 30

for i in data:
    if i == target:
        print("Ditemukan")
```

## **DISKUSI**

**a. Mengapa struktur data penting dalam pemrograman?**

Struktur data penting karena digunakan untuk mengatur dan menyimpan data agar dapat diakses, dikelola, dan diproses secara lebih efisien oleh program.

**b. Apa perbedaan struktur data linear dan non-linear?**

Struktur data linear menyimpan data secara berurutan, seperti array, stack, dan queue. Struktur data non-linear menyimpan data secara tidak berurutan atau bercabang, seperti tree dan graph.

**c. Mengapa algoritma harus efisien?**

Algoritma yang efisien dapat menyelesaikan masalah dengan waktu eksekusi dan penggunaan memori yang lebih sedikit, sehingga program berjalan lebih cepat dan optimal.

**d. Apa fungsi Big-O Notation?**

Big-O Notation digunakan
untuk mengukur dan menggambarkan tingkat efisiensi suatu algoritma berdasarkan pertumbuhan waktu eksekusi atau penggunaan memori terhadap ukuran data.

**e. Mengapa nested loop lebih lambat dibanding single loop?**

Nested loop lebih lambat karena terdapat loop di dalam loop, sehingga jumlah iterasi menjadi jauh lebih banyak dibandingkan single loop. Umumnya nested loop memiliki kompleksitas O(n²), sedangkan single loop memiliki kompleksitas O(n).
