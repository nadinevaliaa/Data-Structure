# **DASAR TEORI**

* **Percabangan (Conditional Statement)**

Percabangan digunakan untuk mengeksekusi blok kode tertentu berdasarkan kondisi yang
dievaluasi. Python menggunakan kata kunci if, elif, dan else. Ekspresi kondisi dievaluasi
sebagai nilai boolean (True/False).

* **Perulangan (Loop)**

Perulangan digunakan untuk mengeksekusi blok kode berulang kali. Python menyediakan
for loop yang berjalan atas suatu iterable (seperti range), dan while loop yang berjalan
selama kondisi bernilai True. Kata kunci break digunakan untuk keluar dari loop,
sedangkan continue untuk melanjutkan ke iterasi berikutnya. Blok else pada loop
dieksekusi jika loop selesai tanpa break.

* **Operator**

Python mendukung berbagai operator: operator aritmatika (+, -, *, /, %), operator
perbandingan (==, !=, >, <, >=, <=), dan operator logika (and, or, not). Operator modulo
(%) sangat berguna untuk memeriksa habis-bagi suatu bilangan.

* **Input/Output**

Fungsi input() digunakan untuk menerima data dari pengguna dalam bentuk string. Fungsi
int() digunakan untuk mengonversi string menjadi integer. Fungsi print() digunakan untuk
menampilkan output ke layar, dengan kemampuan menampilkan beberapa argumen
sekaligus.

```python
# Tugas 1. Menentukan positif, negatif, nol
bil = int(input("Masukkan bilangan: "))

if bil > 0:
  print("Bilangan positif")
elif bil < 0:
  print("Bilangan negatif")
else:
  print("Bilangan nol")
```

```python
# Tugas 4. Bilangan terbesar dari 3 input

a = int(input("Masukkan bilangan 1:"))
b = int(input("Masukkan bilangan 2:"))
c = int(input("Masukkan bilangan 3:"))

if a >= b and a >= c:
  print("Bilangan terbesar:", a)
elif b >= a and b >= c:
  print("Bilangan terbesar:", b)
else:
  print("Bilangan terbesar:", c)
```

## **Analisis:**

Program menerima 3 input integer kemudian membandingkan menggunakan operator logika
AND. Kondisi pertama memeriksa apakah a >= b DAN a >= c. Jika tidak, memeriksa apakah
b >= a DAN b >= c. Jika keduanya tidak, maka c adalah yang terbesar. Dengan input 6, 9, 0.
nilai 9 (b) adalah yang terbesar.

```python
# Tugas 5. Menghitung rata-rata dari 5 nilai yang diinput user

n1 = int(input("Masukkan nilai 1: "))
n2 = int(input("Masukkan nilai 2:"))
n3 = int(input("Masukkan nilai 3: "))
n4 = int(input("Masukkan nilai 4:"))
n5 = int(input("Masukkan nilai 5: "))

rata = (n1 + n2 + n3 + n4 + n5) / 5
print("Rata-rata:", rata)

if rata >= 80:
  print("Baik")
elif rata >= 70:
  print("Cukup")
else:
  print("Kurang")
```

## **Analisis:**

Program menerima 5 input nilai lalu menghitung rata-rata dengan rumus (n1+n2+n3+n4+n5)/5.
Dengan input 40, 90, 80, 77, 100 , rata-rata = 387/5 = 77.4. Karena 77.4 >= 70 tetapi < 80,
maka kategorinya adalah 'Cukup'.

```python
# Tugas 6. Diskon belanja

total = int(input("Masukkan total belanja:"))

if total >= 100000:
  diskon = total * 0.1
else:
  diskon = 0

total_bayar = total - diskon

print("Total belanja:", total_bayar)
print("Diskon", diskon)
```

## **Analisis:**

Program memeriksa apakah total belanja >= 100.000. Jika iya, diskon dihitung sebesar 10%
dari total. Dengan input 455.000, diskon = 455.000 * 0.1 = 45.500, sehingga total_bayar =
455.000 - 45.500 = Rp. 409.500.

```python
# Tugas 7. Menampilkan tabel perkalian dari angka yang diinput user (1-10)

angka = int(input("Masukkan angka:"))

for i in range(1, 11):
  print(angka, "x", i, "=", angka * i)
```

## **Analisis:**

Program menggunakan for loop dengan range(1,11) untuk menghasilkan 10 iterasi (i = 1 sampai
10). Pada setiap iterasi, program mencetak format 'angka x i = hasil'. Dengan input 20, dicetak
tabel perkalian 20 dari 20x1 hingga 20x10.

```python
# Tugas 8. Menghitung faktorial dari suatu bilangan

bil = int(input("Masukkan bilangan: "))
faktorial = 1

for i in range(1, bil + 1):
  faktorial *= i


print("Hasil faktorial:", faktorial)
```

## **Analisis:**

Program dimulai dengan variabel faktorial = 1, lalu nilainya dikalikan secara bertahap dengan
angka dari 1 sampai bilangan yang dimasukkan. Misalnya jika input 8, perhitungannya menjadi
1 × 2 × 3 × 4 × 5 × 6 × 7 × 8 sehingga hasilnya 40320. Operator *= digunakan untuk menyingkat
penulisan, di mana faktorial *= i berarti faktorial = faktorial * i.  

```python
# Tugas 9. Menentukan apakah suatu bilangan termasuk bilangan prima atau bukan

bil = int(input("Masukkan bilangan: "))

if bil <= 1:
    print("Bukan bilangan prima")
else:
    for i in range(2, bil):
        if bil % i == 0:
            print("Bukan bilangan prima")
            break
    else:
        print("Bilangan prima")
```

### **Analisis:**

Program memakai for-else, di mana bagian else akan dijalankan jika perulangan selesai tanpa
break. Saat input 4, program mengecek 4 % 2 == 0 dan hasilnya benar, jadi langsung
menampilkan “Bukan bilangan prima” lalu berhenti karena break. Bilangan prima sendiri hanya
bisa dibagi oleh 1 dan dirinya sendiri.

```python
# Tugas 10. Membuat program login dengan ketentuan:
# Username = "admin", Password = "123"
# Maks percobaan = 3 kali
# Jika gagal 3 kali maka tampilan = "Akses Ditolak"

username_benar = "admin"
password_benar = "123"

for i in range(3):
    username = input("Masukan username: ")
    password = input("Masukan password: ")

    if username == username_benar and password == password_benar:
        print("Login berhasil")
        break
    else:
        print("Password salah, coba lagi")
else:
    print("Akses ditolak")
```

### **Analisis:**

Program menggunakan for-else dengan range(3) untuk memberi batas 3 kali percobaan login.
Jika username dan password benar, program menampilkan “Login berhasil” lalu berhenti
dengan break. Jika semua percobaan gagal dan tidak ada break, bagian else akan dijalankan dan
menampilkan “Akses ditolak”. Ini adalah contoh sederhana sistem autentikasi.
