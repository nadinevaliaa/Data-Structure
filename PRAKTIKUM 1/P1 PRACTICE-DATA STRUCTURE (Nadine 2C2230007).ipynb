## **LATIHAN DI LAB**

```python
# Algoritma Bil Ganjil/Genap

bilangan = int(input("Masukkan Bilangan: "))

if bilangan % 2 == 0:
  print("Bilangan Genap")
else:
  print("Bilangan Ganjil")
```

```python
# Menampilkan Angka 1-10

for i in range(1,11):
  print(i)
```

```python
# Studi Kasus Sederhana (Luas Persegi Panjang)

panjang = int(input("Masukkan Panjang: "))
lebar = int(input("Masukkan Lebar: "))

luas = panjang * lebar

print("Luas persegi panjang:", luas)
```

```python
# Menentukan ganjil/genap (kasus pembagian barang)

jumlah = int(input("Masukkan jumlah buku: "))

if jumlah % 2 == 0:
  print("Jumlah buku genap (bisa dibagi rata)")
else:
  print("Jumlah buku ganjil (tidak bisa dibagi rata)")
```

```python
# Menghitung total belanja

h1 = int(input("Harga barang 1: "))
h2 = int(input("Harga barang 2:"))
h3 = int(input("Harga barang 3: "))

total = h1 + h2 + h3
print("Total belanja:", total)
```

```python
# Menentukan lulus atau tidak

nilai = int(input("Masukkan nilai:"))

if nilai >= 60:
  print("LULUS")
else:
  print("TIDAK LULUS")
```

```python
# Menampilkan nomor antrian

for i in range(1, 11):
  print("Nomor antrian:", i)
```

## **DISKUSI**

### **a. Apa yang dimaksud dengan algoritma?**

Algoritma adalah serangkaian langkah-langkah logis, terstruktur, dan sistematis yang disusun secara runtut untuk menyelesaikan suatu masalah atau menyelesaikan suatu tugas komputasi tertentu.

### **b. Mengapa algoritma penting dalam pemrograman?**

Algoritma bertindak sebagai blueprint atau fondasi logika program. Tanpa algoritma yang matang, kode program akan berantakan, sulit didebug, dan tidak efisien. Algoritma membantu memecahkan masalah besar yang kompleks menjadi instruksi-instruksi kecil yang logis dan terstruktur sehingga komputer dapat mengekskusinya dengan tepat.

### **c. Apa perbedaan algoritma dan program?**

* **Algoritma**: Alur pemikiran, logika, atau langkah penyelesaian masalah. Sifatnya universal dan tidak terikat oleh bahasa pemrograman apa pun (bisa ditulis dalam bentuk teks narasi, pseudocode, atau flowchart).

* **Program**: Implementasi nyata dari suatu algoritma yang sudah ditulis menggunakan sintaks bahasa pemrograman tertentu seperti Python, C++, atau Java agar dapat langsung dijalankan oleh komputer.

### **d. Mengapa Python cocok untuk pemula?**

Python memiliki sintaks yang sederhana, dan menyerupai bahasa Inggris manusia. Pemula tidak perlu dipusingkan dengan aturan penulisan yang rumit seperti wajib menggunakan titik koma ; atau kurung kurawal {} sehingga mereka bisa fokus penuh dalam mempelajari logika berpikir algoritma terlebih dahulu.
