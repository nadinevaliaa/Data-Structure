```python
def tambah (a, b):
  return a + b

hasil = tambah(5, 3)
print("Hasil:", hasil)
```

```python
def hitung_rata(data):
  return sum(data)/len(data)

nilai = []

for i in range(5):
  n = int(input("Masukkan nilai:"))
nilai.append(n)

rata = hitung_rata(nilai)

if rata >=60:
  status ="LULUS"

else:
  status = "TIDAK LULUS"

print("Rata-rata:", rata)
print("Status:", status)
```
