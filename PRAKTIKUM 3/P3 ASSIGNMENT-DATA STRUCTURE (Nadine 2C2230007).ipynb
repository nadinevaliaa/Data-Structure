## **3. DASAR TEORI**

**Struktur Data**

Struktur data adalah cara menyimpan dan mengorganisasikan data di dalam komputer agar dapat diakses dan dikelola secara efisien. Pemilihan struktur data yang tepat dapat meningkatkan kecepatan pemrosesan data serta penggunaan memori. Contoh struktur data yang sering digunakan adalah list, stack, queue, tree, dan graph.

**Abstract Data Type**

Abstract Data Type (ADT) adalah model konseptual yang mendefinisikan jenis data dan operasi yang dapat dilakukan tanpa memperhatikan detail implementasinya. Salah satu contoh ADT adalah stack yang menggunakan prinsip Last In First Out (LIFO) dengan operasi utama push dan pop.

**Efisiensi Algoritma**

Efisiensi algoritma menunjukkan seberapa baik suatu algoritma bekerja dalam menyelesaikan masalah. Algoritma yang efisien dapat memproses data lebih cepat dan menggunakan sumber daya komputer secara optimal, terutama ketika jumlah data yang diolah semakin besar.

**Kompleksitas Algoritma dan Big-O Notation**

Kompleksitas algoritma digunakan untuk mengukur kinerja algoritma berdasarkan waktu eksekusi dan penggunaan memori. Big-O Notation merupakan metode yang digunakan untuk menggambarkan pertumbuhan kompleksitas algoritma terhadap ukuran data masukan. Beberapa kompleksitas yang umum digunakan adalah O(1), O(log n), O(n), dan O(n²).

**Perbandingan Efisiensi Algoritma**

Setiap algoritma memiliki tingkat efisiensi yang berbeda. Sebagai contoh, Linear Search melakukan pencarian data satu per satu, sedangkan Binary Search membagi data menjadi beberapa bagian sehingga proses pencarian dapat dilakukan lebih cepat pada data yang terurut. Oleh karena itu, pemilihan algoritma yang tepat sangat penting untuk meningkatkan performa program.

### **Studi Kasus 1-4**

```python
# STUDI KASUS 1: Penyimpanan Data Mahasiswa

# Membuat list kosong untuk menyimpan nama mahasiswa
mahasiswa = []

# Input jumlah mahasiswa
jumlah = int(input("Masukkan jumlah mahasiswa: "))

# Input nama mahasiswa satu per satu
for i in range(jumlah):
    nama = input(f"Masukkan nama mahasiswa ke-{i+1}: ")
    mahasiswa.append(nama)

# Menampilkan daftar mahasiswa
print("\nDaftar nama mahasiswa:")
for nama in mahasiswa:
    print(f"  - {nama}")
```

### Analisis:

Menggunakan struktur data dynamic list ([]) untuk menampung elemen string secara berurutan. Menggunakan perulangan for berbasis range() untuk menambahkan data secara dinamis via .append().

```python
# STUDI KASUS 2: Simulasi Stack Buku (LIFO)

# Membuat list kosong untuk stack buku
tumpukan_buku = []

# Input jumlah buku
jumlah = int(input("Masukkan jumlah buku: "))

# Menambahkan buku ke stack
for i in range(jumlah):
    buku = input(f"Masukkan nama buku ke-{i+1}: ")
    tumpukan_buku.append(buku)

# Menampilkan isi stack
print("\nBuku ditambahkan:", tumpukan_buku)

# Mengambil buku paling atas (LIFO)
buku_diambil = tumpukan_buku.pop()

print(f"Buku yang diambil (LIFO): {buku_diambil}")
print("Sisa tumpukan:", tumpukan_buku)
```

### Analisis:

Mengimplementasikan konsep Abstract Data Type (ADT) Stack dengan prinsip Last In First Out (LIFO). Operasi append() bertindak sebagai push (menambah ke atas) dan .pop() otomatis mengambil elemen paling terakhir dimasukkan.

```python
# STUDI KASUS 3: Analisis Perulangan (Single vs Nested Loop)

n = 4

# Single loop - O(n)
iterasi_single = 0
print(f"\nSingle Loop (n={n}) - Kompleksitas O(n):")
for i in range(n):
    iterasi_single += 1
    print(f"  i = {i}")
print(f"  Total iterasi: {iterasi_single}")

# Nested loop - O(n^2)
iterasi_nested = 0
print(f"\nNested Loop (n={n}) - Kompleksitas O(n^2):")
for i in range(n):
    for j in range(n):
        iterasi_nested += 1
        print(f"  i={i}, j={j}")
print(f"  Total iterasi: {iterasi_nested}")

print(f"\nKesimpulan: Single loop = {iterasi_single} iterasi, "
      f"Nested loop = {iterasi_nested} iterasi")
```

### Analisis:

Membuktikan perbedaan pertumbuhan Big-O Notation. Single loop berjalan secara linier dengan kompleksitas $O(n)$ , sedangkan nested loop berjalan secara kuadratik dengan kompleksitas $O(n^2)$.

```python
# STUDI KASUS 4: Pencarian Data - Linear Search

daftar_id = [1021, 1034, 1045, 1056, 1067, 1078, 1089]
target_id = 1056

print(f"Daftar ID anggota: {daftar_id}")
print(f"Mencari ID: {target_id}")

ditemukan = False
for id_anggota in daftar_id:
    if id_anggota == target_id:
        ditemukan = True
        break

if ditemukan:
    print(f"  ID {target_id} ditemukan dalam daftar anggota koperasi.")
else:
    print(f"  ID {target_id} tidak ditemukan.")
```

### Analisis:

Menerapkan pencarian sekuensial dengan memeriksa elemen satu per satu dari awal. Kata kunci break digunakan untuk menghentikan proses iterasi segera setelah target_id ditemukan, mengoptimalkan waktu eksekusi.

```python
# Tugas 2 - Simulasi stack sederhana

stack = []

# Push
stack.append(100)
stack.append(200)
stack.append(300)
print("Setelah push 100, 200, 300:", stack)

# Pop
nilai_pop = stack.pop()
print(f"Pop -> nilai yang diambil: {nilai_pop}")
print("Stack setelah pop:", stack)

# Peek (melihat elemen teratas tanpa menghapus)
if stack:
    print(f"Peek: {stack[-1]}")
```

### Analisis:

Selain menggunakan fungsi push (.append()) dan pop (.pop()), terdapat metode peek yang memanfaatkan indeks negatif stack[-1] untuk mengintip elemen teratas tanpa menghapusnya.

```python
# Tugas 3 - Traversal data dengan perulangan

data = [5, 10, 15, 20, 25, 30]
print("Traversal data:")
for index, nilai in enumerate(data):
    print(f"  Index [{index}] -> Nilai: {nilai}")
```

### Analisis:

Melakukan penelusuran list menggunakan fungsi bawaan enumerate(data). Fungsi tsb sangat efisien karena mampu mengembalikan dua nilai sekaligus dalam satu iterasi: indeks posisi dan nilai elemen itu sendiri.

### **Level Sedang**

```python
# Tugas 4 - Menghitung jumlah iterasi dalam loop

ukuran_data = [5, 10, 20, 50]

for n in ukuran_data:
    counter = 0
    for i in range(n):
        counter += 1
    print(f"  n={n:3d} -> jumlah iterasi O(n) = {counter}")
```

### Analisis:

Berdasarkan output, jumlah iterasi yang dilakukan selalu sama dengan nilai n. Saat n = 5, loop berjalan 5 kali, dan ketika n = 50, loop berjalan 50 kali. Hal ini menunjukkan bahwa semakin besar ukuran data, semakin banyak iterasi yang dilakukan secara sebanding, sehingga algoritma memiliki kompleksitas waktu O(n) atau linear.

### **Level Sulit**

```python
# Tugas 8 - Simulasi sistem antrian amhasiswa (Queue)

from collections import deque

class AntrianMahasiswa:
    def __init__(self):
        self.antrian = deque()

    def daftar(self, nama):
        self.antrian.append(nama)
        print(f"  [{nama}] masuk antrian. Posisi ke-{len(self.antrian)}")

    def panggil(self):
        if self.antrian:
            dipanggil = self.antrian.popleft()
            print(f"  Memanggil: [{dipanggil}] | "
                  f"Sisa antrian: {len(self.antrian)} orang")
            return dipanggil
        else:
            print("  Antrian kosong.")
            return None

    def lihat_antrian(self):
        if self.antrian:
            print(f"  Antrian saat ini: {list(self.antrian)}")
        else:
            print("  Antrian kosong.")

antrian = AntrianMahasiswa()

# Mahasiswa mendaftar
print("Mahasiswa mendaftar:")
for nama in ["Reza", "Sinta", "Tono", "Ulfa", "Vino"]:
    antrian.daftar(nama)

print("\nAntrian sebelum dipanggil:")
antrian.lihat_antrian()

print("\nMemanggil mahasiswa satu per satu:")
for _ in range(3):
    antrian.panggil()

print("\nAntrian tersisa:")
antrian.lihat_antrian()
```

### Analisis:

Program mengimplementasikan ADT Queue dengan prinsip First In First Out (FIFO) menggunakan modul collections.deque. Operasi .append() digunakan untuk mengantre di belakang (enqueue), .popleft() digunakan untuk memanggil sekaligus menghapus elemen dari depan antrean (dequeue).
